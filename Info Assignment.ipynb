```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import squarify 
```

```python
wine = pd.read_csv("winequality-red.csv", sep=";")
```

<b>Histogram

```python
plt.figure(figsize=(10, 6))
wine['alcohol'].plot.hist(grid=False, bins=8, rwidth=0.9,color='lightblue')
plt.title("Percentage of wine with its Frequency",fontsize = 20)
plt.xlabel("Percentage of wine",fontsize = 10)
plt.ylabel("Frequency",fontsize =10)
```

<b>Heat Map

```python
correlation = wine.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, linewidths=0, vmin=-1, cmap="RdBu_r")
plt.show()
```

<b>Boxplot

```python
plt.figure(figsize=(10, 6))
ax = sns.boxplot(x="quality", y='alcohol', data = wine)
ax.set_xlabel('Wine Quality',fontsize=10)
ax.set_ylabel('Alcohol Percentage',fontsize=10)
ax.set_title('Alcohol percent in different wine quality types',fontsize=20)
plt.show()
```

<b> Violin Plot

```python
plt.figure(figsize=(10, 6))
bx = sns.violinplot(x="rating", y='citric acid', data = wine)
bx.set_xlabel('Wine Ratings',fontsize=10)
bx.set_ylabel('Citric Acid',fontsize=10)
bx.set_title('Citric acid in different types of Wine ratings',fontsize=20)
plt.show()
```

<b>Swarm Plot

```python
plt.figure(figsize=(10, 6))
cx = sns.swarmplot(x="rating", y="pH", data = wine)
cx.set_xlabel('Wine Ratings',fontsize=10)
cx.set_ylabel('pH',fontsize=10)
cx.set_title('pH in different types of Wine ratings',fontsize=20)
plt.show()
```

<b>lmplot

```python
sns.lmplot(x = "alcohol", y = "residual sugar", col = "rating", data = wine)
plt.show()
```

<b>Pie Chart

```python
labels = ['midsize', 'compact', 'minivan', '2seater','suv']
sizes = [15, 30, 45, 10,27]
explode = (0, 0, 0.1, 0,0)  
color = ["lightblue","lightyellow","khaki","gray","lavender"]
fig1, ax1 = plt.subplots(figsize=(10,6))
ax1.pie(sizes, explode=explode, labels=labels, colors = color,autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  
plt.tight_layout()
plt.title("Vehicle Class Distribution",fontsize = 20)
plt.show()
```

<b>Treemap

```python
df_raw = pd.read_csv("car.csv")
df = df_raw.groupby('class').size().reset_index(name='counts')
```

```python
labels = df.apply(lambda x: str(x[0]) + "\n (" + str(x[1]) + ")", axis=1)
sizes = df['counts'].values.tolist()
colors = [plt.cm.Spectral(i/float(len(labels))) for i in range(len(labels))]

# Draw Plot
plt.figure(figsize=(12,8), dpi= 80)
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=.8)

# Decorate
plt.title('Treemap of Vechile Class')
plt.axis('off')
plt.show()
```

<b>Scatter Plot

```python
iris = pd.read_csv("iris.csv")
```

```python
# create a figure and axis
fig, ax = plt.subplots(figsize=(10,6))
# scatter the sepal_length against the sepal_width
ax.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'])
# set a title and labels
ax.set_title('Sepal Length vs Sepal Width',fontsize = 20 )
ax.set_xlabel('sepal length')
ax.set_ylabel('sepal width')
plt.show()
```

<b>Line Chart

```python
columns = iris.columns.drop(['Species','Id'])
x_data = range(0, iris.shape[0])
fig, ax = plt.subplots(figsize=(10,6))

for column in columns:
    ax.plot(x_data, iris[column])
ax.set_title('Iris Dataset')
ax.legend()
```

<b>Bar Chart

```python
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Spark','C')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,3,2,1]
plt.figure(figsize=(10, 6))
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.xlabel('Language')
plt.ylabel('Usage')
plt.title('Programming language usage')
plt.show()
```
